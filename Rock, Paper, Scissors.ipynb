```python
import random as rand

print('The "Rock, Paper, Scissors" Program')
print()

while (True):

    user_str = input('Enter 1 for "Rock", 2 for "Paper", or 3 for "Scissors": ')
    try:
        user = int(user_str)
        user_go = True
        if (user < 1 or user > 3):
            print ("Invalid entry, the choice must be between 1 and 3.")
            print ()
            user_go = False

    except:
        print ("Invalid entry, you must enter an integer")
        print ()
        user_go = False
    
    if (user_go) :
        print ()
        computer = rand.randint(1,3)
        if (user == computer) :
            print ("It's a draw.")
        elif ((user == 1 and computer == 3) or (user == 2 and computer == 1) or (user == 3 and computer == 2)) :
        # Note: There other ways of programming this to take this branch randomly 50% of the remaining time,
        #   but the condition clause above is most closely analogous to the terms of the game.
            print ("You win!")
        else :
            print ("You lose.")
        print ()
     
    again = input("Do you want to play again? (y for yes, anything else for no) ")
    print ()
    if (again != "y") :
        break

print ("Thank you for playing the game!")
```
