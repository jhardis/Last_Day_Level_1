```python
print('The "DataWork Cafe" Program')
print()

menu = {}    # Note:  This is a glorified two-dimensional array
menu[1] = {"category": "Coffee", "item": "Expresso", "price": 1.50, "qty": 0}
menu[2] = {"category": "Coffee", "item": "Americano", "price": 2.00, "qty": 0}
menu[3] = {"category": "Coffee", "item": "Caffe latte", "price": 2.25, "qty": 0}
menu[4] = {"category": "Coffee", "item": "Mocha", "price": 2.25, "qty": 0}
menu[5] = {"category": "Coffee", "item": "Cappucino", "price": 2.50, "qty": 0}
menu[6] = {"category": "Coffee", "item": "Macchiato", "price": 2.50, "qty": 0}
menu[7] = {"category": "Tea", "item": "White tea", "price": 1.50, "qty": 0}
menu[8] = {"category": "Tea", "item": "Green tea", "price": 2.00, "qty": 0}
menu[9] = {"category": "Tea", "item": "Oolong tea", "price": 2.25, "qty": 0}
menu[10] = {"category": "Tea", "item": "Black tea", "price": 2.25, "qty": 0}
menu[11] = {"category": "Tea", "item": "Earl Grey", "price": 2.50, "qty": 0}
menu[12] = {"category": "Tea", "item": "Matcha", "price": 2.50, "qty": 0}
menu[13] = {"category": "Pastries", "item": "Croissant", "price": 5.50, "qty": 0}
menu[14] = {"category": "Pastries", "item": "Cinnamon roll", "price": 4.50, "qty": 0}
menu[15] = {"category": "Pastries", "item": "Banana bread", "price": 4.25, "qty": 0}
menu[16] = {"category": "Pastries", "item": "Pumpkin bread", "price": 3.75, "qty": 0}
menu[17] = {"category": "Pastries", "item": "Sticky bun", "price": 4.25, "qty": 0}
menu[18] = {"category": "Pastries", "item": "Apple turnover", "price": 3.75, "qty": 0}

print("Welcome to the MC Cafe!")
print()

ans = input("Would you like to see the menu? (y for yes, anything else for no) ")
print()

if (ans == "y"):
    header = ""
    for i in range(1,19):
        offer = menu[i]
        if (offer["category"] != header) :
            header = offer["category"]
            print (f"• {header}")
        print (f"   o {offer['item']} - ${offer['price']:.2f}")
    print ()

ans = input("Would you like to order? (y for yes, anything else for no) ")

if (ans == "y"):

    while (True) :       # Loop to take order
        print()
        ans = input("What would you like to order? (Spelling matters!) ")
        found = 0
        for i in range(1,19) :
            item = menu[i]["item"]
            if (menu[i]["item"].lower() == ans.lower()) :
                found = i
                break
        if (found == 0) :
            print ("Sorry, no such item.")
        else :
            count = -1
            while (count < 0) :
                ans = input(f"How many {menu[i]['item']} would you like to order? ")
                try:
                    count1 = int(ans)
                except:
                    print("Sorry, invalid entry.")
                    print()
                    continue
                if (count1 < 0) :
                    print("Sorry, invalid entry.")
                    print()
                    continue
                else :
                    count = count1
            menu[found]["qty"] = count

        print()
        ans = input("Would you like to order another item? (y for yes, anything else for no) ")
        if (ans != "y") :
            break
        # End of loop to take order

    # Display an order receipt
    print ()
    print ()
    print ("            RECEIPT")
    print ()
    total = 0
    for i in range(1,19) :
        if (menu[i]["qty"] != 0) :
            item = menu[i]['item']
            padding = 22-len(item)
            j = menu[i]["qty"]
            while (j > 0) :
                padding -= 1
                j //= 10
            print(f"{item}{' '*padding}{menu[i]['qty']} @ ${menu[i]['price']:.2f}")
            total += menu[i]["qty"] * menu[i]["price"]
    
    print ("="*30)
    
    padding = 17
    j = total
    while (j > 10) :
        padding -= 1
        j /= 10
    print (f"Subtotal{' '*padding}${total:.2f}")
    
    tax = round(0.06*total,2)
    padding = 11
    j = tax
    while (j > 10) :
        padding -= 1
        j /= 10
    print (f"Sales Tax (6%){' '*padding}${tax:.2f}")
    
    print ("="*30)
    
    padding = 14
    j = total+tax
    while (j > 10) :
        padding -= 1
        j /= 10
    print (f"Grand Total{' '*padding}${total+tax:.2f}")
    print ()
    # End of block to print the order receipt
    
    # End of "Would like to Order" block

print()
print("Thank you for visiting the MC Cafe!")
```
